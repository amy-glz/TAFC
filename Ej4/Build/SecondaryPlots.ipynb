```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("secondary.csv")
# Conteo por tipo de part√≠cula
print(df["name"].value_counts())
```

```python
print(df.columns)
```

```python
# Histograma de energ√≠as
df[df["name"]=="e-"]["ekin_MeV"].hist(bins=50)
plt.xlabel("Energy [MeV]")
plt.ylabel("Counts")
plt.title("Electron energy spectrum")
plt.show()
mom = df[["momX","momY","momZ"]].to_numpy()
cos_theta = mom[:,2] / np.linalg.norm(mom, axis=1)
plt.hist(cos_theta, bins=50)
plt.xlabel("cos(theta)")
plt.ylabel("Counts")
plt.title("Angular distribution")
plt.show()
```
