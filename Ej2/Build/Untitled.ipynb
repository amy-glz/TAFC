```python
import numpy as np
import matplotlib.pyplot as plt
import pandas 
```

```python
data = np.loadtxt("dDep.txt", comments="#", delimiter=',')
# columnas: [ir, iphi, iz, dose, error]
x, y, z = data[:,0], data[:,1], data[:,2]
dose = data[:,3]

side=50
size=max(x)

x1=np.linspace(0,size,11)
x2=np.linspace(-side/2,side/2,11)
for e in range(len(x2)): 
    x2[e]=(x2[e])

print(x1)
print(x2)
```

```python
xi = x.astype(int)
yi = y.astype(int)
nx = np.max(xi) + 1
ny = np.max(yi)+ 1
mapa_dosis = np.zeros((ny, nx))
for i in range(len(dose)):
    mapa_dosis[yi[i], xi[i]] += dose[i]
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(mapa_dosis, origin='lower', cmap='jet', interpolation='nearest',)  
plt.colorbar(label='Dosis absorbida [Gy]')
plt.title('Mapa de dosis absorbida, número de vóxel = '+ str(int(size)))
plt.xlabel('X [mm]')
plt.xticks(x1,x2)
plt.ylabel('Y [mm]')
plt.yticks(x1,x2)
#plt.legend()
plt.savefig("MapaDosisVS" + str(size) + ".png",transparent=None, dpi='figure', format="png")
plt.show()
```
