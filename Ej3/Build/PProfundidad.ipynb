```python
import numpy as np
import matplotlib.pyplot as plt
import pandas 
```

```python
data = np.loadtxt("dDep.txt", comments="#", delimiter=',')
# columnas: [ir, iphi, iz, dose, error]
x, y, z = data[:,0], data[:,1], data[:,2]
dose = data[:,3]

side=50
size=max(z)
aa=11

x1=np.linspace(0,size,aa)
x2=np.linspace(0,side,aa)
for e in range(len(x2)): 
    x2[e]=(x2[e])

dmax=max(dose)
pdd=(dose/dmax)*100
```

```python
plt.figure(figsize=(8, 6))
plt.plot(z,dose,'--o')  
#plt.colorbar(label='Dosis absorbida [Gy]')
plt.title('Perfil de dosis lateral Gammas de 6 MeV')
plt.xlabel('Profundidad en z [mm]')
plt.xticks(x1,x2)
plt.ylabel('Dosis [Gy]')
#plt.yticks(x1,x2)
#plt.legend()
plt.savefig("Perfil de dosis lateral Gammas 6 MeV.png",transparent=None, dpi='figure', format="png")
plt.show()
```
